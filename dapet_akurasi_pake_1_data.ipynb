```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan FFT dan mengembalikan frekuensi serta magnitude
def plot_fft(signal, sampling_rate, channel_name):
    # Melakukan FFT
    N = len(signal)  # Panjang sinyal
    T = 1.0 / sampling_rate  # Interval sampling
    fft_vals = np.fft.fft(signal)  # Menghitung FFT
    fft_freqs = np.fft.fftfreq(N, T)  # Frekuensi
    fft_vals = np.abs(fft_vals[:N // 2])  # Mengambil magnitude frekuensi positif saja
    fft_freqs = fft_freqs[:N // 2]  # Mengambil frekuensi positif saja

    # Plot domain frekuensi
    plt.figure(figsize=(10, 6))
    plt.plot(fft_freqs, fft_vals)
    plt.title(f'FFT of {channel_name} Channel')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()

# Contoh penggunaan untuk beberapa saluran EEG
sampling_rate = 256  # Contoh sampling rate 256 Hz, bisa disesuaikan sesuai data

# Menampilkan domain frekuensi untuk beberapa saluran EEG
channels = ['FT10-T8', 'FT9-FT10', 'F8-T8']  # Pilih beberapa saluran

for channel in channels:
    signal = data1[channel].values  # Mengambil sinyal dari setiap saluran
    plot_fft(signal, sampling_rate, channel)  # Memanggil fungsi FFT untuk setiap saluran
```

```python
import numpy as np

# Fungsi untuk mendapatkan frekuensi terbesar dari FFT
def get_max_frequency(signal, sampling_rate):
    N = len(signal)
    T = 1.0 / sampling_rate
    fft_vals = np.fft.fft(signal)
    fft_freqs = np.fft.fftfreq(N, T)
    fft_vals = np.abs(fft_vals[:N // 2])
    fft_freqs = fft_freqs[:N // 2]

    max_freq_index = np.argmax(fft_vals)
    return fft_freqs[max_freq_index], fft_vals[max_freq_index]

# Sampling rate
sampling_rate = 256  # Misal 256 Hz

# Membuat dictionary untuk menyimpan hasil
max_frequencies = {}

# Menganalisis setiap saluran EEG
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3', 'C3-P3', 'P3-O1',
            'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2',
            'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Loop untuk menghitung frekuensi terbesar untuk setiap saluran
for channel in channels:
    signal = df2[channel].values  # Ambil sinyal dari setiap saluran EEG
    max_freq, max_magnitude = get_max_frequency(signal, sampling_rate)  # Dapatkan frekuensi dan magnitude terbesar
    max_frequencies[channel] = (max_freq, max_magnitude)  # Simpan hasil ke dictionary

# Menampilkan hasil saluran dengan frekuensi terbesar
sorted_max_freq = sorted(max_frequencies.items(), key=lambda x: x[1][1], reverse=True)

# Menampilkan saluran dengan frekuensi terbesar
print("Saluran EEG dengan frekuensi terbesar yang mungkin menyebabkan kejang:")
for channel, (freq, mag) in sorted_max_freq:
    print(f"Channel: {channel}, Max Frequency: {freq:.2f} Hz, Magnitude: {mag:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan FFT dan mengembalikan frekuensi serta magnitude
def compute_fft(signal, sampling_rate):
    N = len(signal)  # Panjang sinyal
    T = 1.0 / sampling_rate  # Interval sampling
    fft_vals = np.fft.fft(signal)  # Menghitung FFT
    fft_freqs = np.fft.fftfreq(N, T)  # Frekuensi
    fft_vals = np.abs(fft_vals[:N // 2])  # Magnitude (frekuensi positif saja)
    fft_freqs = fft_freqs[:N // 2]  # Frekuensi positif saja
    return fft_freqs, fft_vals

# Sampling rate
sampling_rate = 256  # Misal 256 Hz

# Pilih beberapa atau semua saluran EEG untuk ditampilkan
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3', 'C3-P3', 'P3-O1',
            'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2',
            'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Membuat plot untuk semua saluran EEG dalam satu gambar
plt.figure(figsize=(14, 8))

for channel in channels:
    signal = df2[channel].values  # Mengambil sinyal dari saluran EEG
    fft_freqs, fft_vals = compute_fft(signal, sampling_rate)  # Menghitung FFT
    plt.plot(fft_freqs, fft_vals, label=channel)  # Menambahkan ke plot

# Memberikan judul dan label pada plot
plt.title('FFT of All EEG Channels')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.legend(loc='upper right')  # Menampilkan legenda untuk setiap saluran
plt.grid(True)
plt.show()
```

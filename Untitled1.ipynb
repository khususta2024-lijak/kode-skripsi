```python
# ini random forest
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the data from the CSV file into a pandas DataFrame
TA_27 = pd.read_csv("/content/drive/MyDrive/TA/SEMUA DATA YANG DIBENERI LABEL NYA.csv")

# Step 11: Siapkan data untuk klasifikasi
X = TA_27.drop(['frekuensi', 'Label'], axis=1)
y = TA_27['Label']

# Step 12: Bagi data menjadi latih dan uji secara stratified
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)

# Step 13: Latih model Random Forest dengan max_features='sqrt'
model = RandomForestClassifier(n_estimators=50, max_depth=10, max_features='sqrt', random_state=42)
model.fit(X_train, y_train)

# Prediksi pada data latih
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Prediksi pada data uji
y_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)

# Step 14: Evaluasi model -> Cetak akurasi dan laporan
print("Akurasi pada data latih:", train_accuracy)  # Akurasi pada data latih
print("Akurasi pada data uji:", test_accuracy)  # Akurasi pada data uji
print("Laporan Klasifikasi (Data Uji):\n", classification_report(y_test, y_pred))

# Step 15: Matriks Evaluasi
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Tidak Kejang', 'Kejang'], yticklabels=['Tidak Kejang', 'Kejang'])
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca data EEG (yang sudah diberi label) dari CSV
df = pd.read_csv("/content/drive/MyDrive/TA/LABEL/chb15_17.csv")

channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3',
            'C3-P3', 'P3-O1', 'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8',
            'F8-T8', 'T8-P8-0', 'P8-O2', 'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9',
            'FT9-FT10', 'FT10-T8', 'T8-P8-1']
sampling_rate = 256  # Hz
window_size = 1      # dalam detik
samples_per_window = window_size * sampling_rate  # Jumlah sampel per segmen

# Fungsi untuk menghitung FFT dari suatu segmen sinyal
def compute_fft(signal, sampling_rate):
    N = len(signal)
    T = 1.0 / sampling_rate
    fft_vals = np.fft.fft(signal)
    fft_freqs = np.fft.fftfreq(N, T)
    # Ambil setengah spektrum (frekuensi positif)
    fft_vals = np.abs(fft_vals[:N // 2])
    fft_freqs = fft_freqs[:N // 2]
    return fft_freqs, fft_vals

fft_features = []  # Menyimpan fitur FFT untuk tiap segmen (rata-rata dari semua channel)
labels = []        # Menyimpan label tiap segmen
frequency_axis = None  # Akan diisi dengan sumbu frekuensi (sama untuk setiap segmen)

# Proses data berdasarkan segmen
for start in range(0, len(df), samples_per_window):
    end = start + samples_per_window
    segment = df.iloc[start:end]
    # Pastikan segmen cukup panjang
    if len(segment) != samples_per_window:
        continue

    # Ambil label segmen: contoh, gunakan aturan label dari kolom 'seizure'
    # Misalnya, jika ada minimal satu data dengan label "Seizure", maka segmen diberi label "Seizure"
    seg_label = "Seizure" if any(segment['seizure'] == "Seizure") else "Tidak Kejang"
    labels.append(seg_label)

    # Hitung FFT untuk tiap channel dalam segmen dan simpan hasilnya
    segment_fft_list = []
    for channel in channels:
        if channel in df.columns:
            signal_seg = segment[channel].values
            fft_freqs, fft_vals = compute_fft(signal_seg, sampling_rate)
            segment_fft_list.append(fft_vals)

    # Jika ada data FFT dari channel, hitung rata-rata (atau Anda bisa menyimpan semua secara terpisah)
    if segment_fft_list:
        avg_fft = np.mean(segment_fft_list, axis=0)
        fft_features.append(avg_fft)
        # Simpan sumbu frekuensi (akan sama untuk semua segmen)
        if frequency_axis is None:
            frequency_axis = fft_freqs

# Buat DataFrame dari hasil FFT dan sertakan label untuk tiap segmen
TA_fft = pd.DataFrame(fft_features)
TA_fft['seizure'] = labels
# Jika diinginkan, simpan sumbu frekuensi sebagai kolom berupa list
TA_fft['frekuensi'] = [list(frequency_axis)] * TA_fft.shape[0]

# Contoh visualisasi: Plot FFT segmen pertama untuk beberapa channel
plt.figure(figsize=(16, 10))
plt.plot(frequency_axis, TA_fft.iloc[0, :-2], label='Rata-rata FFT Segmen 1')
plt.title('FFT EEG Data (Segmen 1)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft

# Membaca data EEG (yang sudah diberi label) dari CSV
df = pd.read_csv("/content/drive/MyDrive/TA/LABEL/chb15_17.csv")

channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3',
            'C3-P3', 'P3-O1', 'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8',
            'F8-T8', 'T8-P8-0', 'P8-O2', 'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9',
            'FT9-FT10', 'FT10-T8', 'T8-P8-1']

df['seizure'] = df['seizure'].astype(int)  # Ubah kolom label ke integer

# Pastikan df sudah ada dengan kolom ['time', 'eeg_signal', 'label']
fs = 256  # Sampling rate (Hz)
window_size = fs  # 1 detik = 256 sampel
num_windows = len(df) // window_size

features = []
labels_freq = []

for i in range(num_windows):
    # Ambil segmen sinyal EEG dan label
    segment = df[channels].iloc[i * window_size:(i + 1) * window_size].values
    segment_label = df['seizure'].iloc[i * window_size:(i + 1) * window_size].values

    # Lakukan FFT
    fft_values = np.abs(fft(segment))[:window_size // 2]  # Ambil frekuensi positif saja
    frequencies = np.fft.fftfreq(window_size, d=1/fs)[:window_size // 2]

    # Majority Labeling ( 50% sampel dalam segmen adalah kejang  label kejang)
    seizure_count = np.sum(segment_label)
    label = 1 if seizure_count >= (window_size // 2) else 0

    features.append(fft_values)
    labels_freq.append(label)

# Konversi ke DataFrame
df_fft = pd.DataFrame(features)
df_fft['label'] = labels_freq

# Tampilkan hasil
print(df_fft.head())  # Cek hasil ekstraksi fitur FFT dan label

# Plot spektrum rata-rata kejang vs tidak kejang
plt.figure(figsize=(10,5))
plt.plot(frequencies, np.mean(df_fft[df_fft['label'] == 0].iloc[:, :-1], axis=0), label="Tidak Kejang", color='blue')
plt.plot(frequencies, np.mean(df_fft[df_fft['label'] == 1].iloc[:, :-1], axis=0), label="Kejang", color='red')
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Amplitudo FFT")
plt.title("Spektrum EEG Kejang vs Tidak Kejang")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft

# Baca file CSV dengan delimiter yang benar
df = pd.read_csv("/content/drive/MyDrive/TA/LABEL/chb15_17.csv")  # Sesuaikan delimiter jika perlu

# Hapus spasi dari nama kolom jika ada
df.columns = df.columns.str.strip()

# Pastikan kolom 'seizure' berupa angka (0 = tidak kejang, 1 = kejang)
df['seizure'] = df['seizure'].map({'tidak kejang': 0, 'kejang': 1})

# Parameter
fs = 256  # Sampling rate (Hz)
window_size = fs  # 1 detik = 256 sampel
num_windows = len(df) // window_size  # Jumlah segmen

# Ambil hanya kolom EEG (kecuali 'Time' dan 'seizure')
eeg_channels = [col for col in df.columns if col not in ['Time', 'seizure']]

features = []
labels_freq = []

for i in range(num_windows):
    # Ambil segmen sinyal EEG untuk setiap channel
    segment = df.iloc[i * window_size:(i + 1) * window_size][eeg_channels].values
    segment_label = df['seizure'].iloc[i * window_size:(i + 1) * window_size].values

    # Lakukan FFT pada setiap channel EEG
    fft_values = np.abs(fft(segment, axis=0))[:window_size // 2]  # Ambil frekuensi positif
    frequencies = np.fft.fftfreq(window_size, d=1/fs)[:window_size // 2]

    # Majority Labeling ( 50% sampel dalam segmen adalah kejang  label kejang)
    seizure_count = np.sum(segment_label)
    label = 1 if seizure_count >= (window_size // 2) else 0

    # Simpan fitur FFT rata-rata tiap channel
    fft_features = {channel: fft_values[:, idx].mean() for idx, channel in enumerate(eeg_channels)}
    fft_features['seizure'] = label
    features.append(fft_features)

# Konversi hasil ke DataFrame
df_fft = pd.DataFrame(features)

# Tampilkan hasil
print(df_fft.head())  # Cek hasil ekstraksi fitur FFT

# Visualisasi perbandingan kejang vs tidak kejang pada salah satu channel EEG
channel_to_plot = eeg_channels[0]  # Misalnya FP1-F7
plt.figure(figsize=(10, 5))
plt.hist(df_fft[df_fft['seizure'] == 0][channel_to_plot], bins=30, alpha=0.7, label="Tidak Kejang", color='blue')
plt.hist(df_fft[df_fft['seizure'] == 1][channel_to_plot], bins=30, alpha=0.7, label="Kejang", color='red')
plt.xlabel("Amplitudo FFT (Mean)")
plt.ylabel("Frekuensi")
plt.title(f"Distribusi FFT pada Channel {channel_to_plot}")
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.signal as signal
import pandas as pd

# Load data EEG
TA_1 = pd.read_csv("/content/drive/MyDrive/TA/FILTERING/chb05_06.csv")

# Sampling rate EEG
sampling_rate = 256

# Fungsi untuk menerapkan FIR Band-Pass Filter dengan Hamming window
def apply_fir_bandpass(signal_data, lowcut, highcut, sampling_rate, num_taps=101):
    nyquist = 0.5 * sampling_rate  # Nyquist frequency
    low = lowcut / nyquist
    high = highcut / nyquist

    # Buat koefisien filter dengan Hamming window
    fir_coeff = signal.firwin(num_taps, [low, high], pass_zero=False, window="hamming", fs=sampling_rate)

    # Terapkan filter ke sinyal EEG
    filtered_signal = signal.lfilter(fir_coeff, 1.0, signal_data)

    return filtered_signal

# Pilih channel EEG untuk diproses (tanpa kolom Time & seizure)
channels = [col for col in TA_1.columns if col not in ['Time', 'seizure']]

# Buat DataFrame untuk hasil filtering
filtered_data_kejang = pd.DataFrame()
filtered_data_tidak_kejang = pd.DataFrame()

for channel in channels:
    # Filtering 1-25 Hz untuk data kejang
    filtered_data_kejang[channel] = apply_fir_bandpass(TA_1[channel].values, 1, 25, sampling_rate)

    # Filtering 30-50 Hz untuk data tidak kejang
    filtered_data_tidak_kejang[channel] = apply_fir_bandpass(TA_1[channel].values, 30, 50, sampling_rate)

# Tambahkan label
filtered_data_kejang['seizure'] = 1
filtered_data_tidak_kejang['seizure'] = 0

# Gabungkan data hasil filtering
filtered_data = pd.concat([filtered_data_kejang, filtered_data_tidak_kejang])

# Cek distribusi label
print(filtered_data['seizure'].value_counts())

# Simpan hasil filtering jika diperlukan
filtered_data.to_csv("/content/drive/MyDrive/TA/FILTERING/filtered_eeg_1-25_30-50.csv", index=False)
```

```python
def konvolusi(x, h):
  N = len(x)
  M = len(h)

  panjang = N + M -1
  y = [0] * panjang

  for n in range(panjang):
    for k in range(N):
      if 0 <= n-k < M:
        y[n] += x[k] * h[n-k]

  return y[:N]


filtering = konvolusi(x, taps)
print(filtering)
```

```python
import numpy as np

def dft(x):
    """
    Menghitung Discrete Fourier Transform (DFT) secara manual
    Input: x (list atau numpy array) = sinyal domain waktu
    Output: X (numpy array kompleks) = spektrum frekuensi
    """
    N = len(x)
    X = []
    for k in range(N):  # looping frekuensi
        s = 0
        for n in range(N):  # looping waktu
            faktor = np.exp(-2j * np.pi * k * n / N)  # faktor kompleks
            s += x[n] * faktor
        X.append(s)
    return np.array(X)

# contoh sinyal
x = [1, 2, 3, 4]

# hitung DFT
X = dft(x)

# tampilkan hasil
print("Input:", x)
for k, val in enumerate(X):
    print(f"X[{k}] = {val:.4f}")
```

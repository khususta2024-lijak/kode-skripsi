# MAX AMPLITUDO (RENTANG FREKUENSI 0-1, 1-2, 2-3)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("/content/drive/MyDrive/TA/FFT YANG DIPISAH (KEJANG).csv")

# Menghapus kolom non-channel (seizure dan frekuensi)
channel_columns = [col for col in data.columns if col not in ["seizure", "frekuensi"]]

# Tentukan rentang frekuensi 0-50 Hz dengan interval 1 Hz
rentang_frekuensi = list(range(0, 51))

# Buat dataframe untuk menyimpan hasil
hasil = pd.DataFrame(columns=["Rentang Frekuensi", "Amplitudo Maksimum"])

# Loop melalui setiap rentang frekuensi
temp = []
for i in range(len(rentang_frekuensi) - 1):
    lower_bound = rentang_frekuensi[i]
    upper_bound = rentang_frekuensi[i + 1]

    # Filter data berdasarkan rentang frekuensi
    subset = data[(data["frekuensi"] >= lower_bound) & (data["frekuensi"] < upper_bound)]

    # Ambil amplitudo maksimum di semua channel dalam rentang tersebut
    max_amplitude = subset[channel_columns].max().max()

    # Tambahkan hasil ke list
    temp.append([f"{lower_bound}-{upper_bound} Hz", max_amplitude])

# Konversi list ke dataframe
hasil = pd.DataFrame(temp, columns=["Rentang Frekuensi", "Amplitudo Maksimum"])

# Menampilkan hasil
print(hasil)

# Visualisasi histogram amplitudo maksimum
plt.figure(figsize=(12, 6))
plt.bar(hasil["Rentang Frekuensi"], hasil["Amplitudo Maksimum"], color='blue', alpha=0.7)
plt.xlabel("Rentang Frekuensi (Hz)")
plt.ylabel("Amplitudo Maksimum")
plt.title("Histogram Amplitudo Maksimum di Seluruh Channel per Rentang Frekuensi")
plt.xticks(rotation=90)
plt.show()

# Simpan data yang digabung
hasil.to_csv("/content/drive/MyDrive/TA/AMPLITUDO TERTINGGI (KEJANG).csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "/content/drive/MyDrive/TA/FFT YANG DIPISAH (KEJANG).csv"
data = pd.read_csv(file_path)

# Pastikan kolom yang berisi frekuensi dan amplitudo sesuai dengan dataset Anda
frequencies = data['frekuensi']

# Daftar channel EEG
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3',
            'C3-P3', 'P3-O1', 'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8',
            'T8-P8-0', 'P8-O2', 'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Definisikan rentang frekuensi masing-masing gelombang
bands = {
    "Delta": (1, 4),
    "Theta": (4, 8),
    "Alpha": (8, 13),
    "Beta": (13, 30),
    "Gamma": (30, 50)
}

def get_band_power(frequencies, amplitudes, band):
    mask = (frequencies >= band[0]) & (frequencies < band[1])
    return np.sum(amplitudes[mask]**2)

# Hitung daya spektral untuk masing-masing channel dan gelombang
channel_band_powers = {}
for channel in channels:
    amplitudes = data[channel]  # Ambil data amplitudo per channel
    band_powers = {band: get_band_power(frequencies, amplitudes, bands[band]) for band in bands}
    channel_band_powers[channel] = band_powers

# Tampilkan hasil per channel
for channel, powers in channel_band_powers.items():
    print(f"\nChannel: {channel}")
    for band, power in powers.items():
        print(f"  Daya {band}: {power:.4f}")

# Visualisasi hasil untuk semua channel
plt.figure(figsize=(12, 6))
for band in bands:
    values = [channel_band_powers[channel][band] for channel in channels]
    plt.plot(channels, values, marker='o', label=band)

plt.xticks(rotation=90)
plt.xlabel("Channel EEG")
plt.ylabel("Daya Spektral")
plt.title("Daya Spektral per Gelombang di Setiap Channel")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data 1 (Kejang)
file_path1 = "/content/drive/MyDrive/TA/FFT YANG DIPISAH (KEJANG).csv"
data1 = pd.read_csv(file_path1)

# Load data 2 (Normal)
file_path2 = "/content/drive/MyDrive/TA/FFT YANG DIPISAH (NO KEJANG).csv"
data2 = pd.read_csv(file_path2)

# Pastikan kolom yang berisi frekuensi dan amplitudo sesuai dengan dataset Anda
frequencies1 = data1['frekuensi'].values
frequencies2 = data2['frekuensi'].values

# Daftar channel EEG
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3',
            'C3-P3', 'P3-O1', 'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2',
            'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2', 'FZ-CZ', 'CZ-PZ',
            'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Definisikan rentang frekuensi masing-masing gelombang
bands = {
    "Delta": (1, 4),
    "Theta": (4, 8),
    "Alpha": (8, 13),
    "Beta": (13, 30),
    "Gamma": (30, 50)
}

# Fungsi untuk menghitung PSD menggunakan metode periodogram
def compute_psd(frequencies, amplitudes, band):
    mask = (frequencies >= band[0]) & (frequencies < band[1])
    psd_values = (amplitudes[mask]**2) / len(amplitudes)  # PSD = |X(f)| / N
    return np.sum(psd_values)

# Hitung PSD total dari seluruh channel EEG
total_amplitudes1 = data1[channels].sum(axis=1).values
total_amplitudes2 = data2[channels].sum(axis=1).values

psd_values1 = {band: compute_psd(frequencies1, total_amplitudes1, bands[band]) for band in bands}
psd_values2 = {band: compute_psd(frequencies2, total_amplitudes2, bands[band]) for band in bands}

# Tampilkan hasil
for band, psd in psd_values1.items():
    print(f"PSD {band} (Kejang): {psd:.4f}")
for band, psd in psd_values2.items():
    print(f"PSD {band} (Normal): {psd:.4f}")

# Visualisasi hasil
plt.figure(figsize=(8, 5))
bar_width = 0.4
index = np.arange(len(bands))

plt.bar(index, psd_values1.values(), bar_width, label='Kejang', color='blue')
plt.bar(index + bar_width, psd_values2.values(), bar_width, label='Normal', color='red')

plt.xlabel("Jenis Gelombang")
plt.ylabel("Power Spectral Density (PSD)")
plt.title("Perbandingan PSD Gelombang (Kejang vs Normal)")
plt.xticks(index + bar_width / 2, bands.keys())
plt.legend()
plt.show()
```

# STATISTIK DESKRIPTIF MAX AMPLITUDO (FREKUENSI)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
kejang = pd.read_csv("/content/drive/MyDrive/TA/AMPLITUDO TERTINGGI (KEJANG).csv")
no_kejang = pd.read_csv("/content/drive/MyDrive/TA/AMPLITUDO TERTINGGI (NO KEJANG).csv")

# Pastikan kolom Rentang Frekuensi dan Amplitudo Maksimum ada
def validate_columns(df, name):
    required_columns = {"Rentang Frekuensi", "Amplitudo Maksimum"}
    missing_columns = required_columns - set(df.columns)
    if missing_columns:
        raise KeyError(f"Kolom {missing_columns} tidak ditemukan dalam dataset {name}")

validate_columns(kejang, "kejang")
validate_columns(no_kejang, "no_kejang")

# Menghitung rasio amplitudo kejang terhadap no kejang
merged_df = kejang.merge(no_kejang, on="Rentang Frekuensi", suffixes=("_kejang", "_no_kejang"))
merged_df["Rasio Amplitudo"] = merged_df["Amplitudo Maksimum_kejang"] / merged_df["Amplitudo Maksimum_no_kejang"]

# Visualisasi histogram amplitudo maksimum untuk kejang dan no kejang
plt.figure(figsize=(12, 6))
plt.bar(kejang["Rentang Frekuensi"], kejang["Amplitudo Maksimum"], color='red', alpha=0.7, label='Kejang')
plt.bar(no_kejang["Rentang Frekuensi"], no_kejang["Amplitudo Maksimum"], color='blue', alpha=0.7, label='No Kejang')
plt.xlabel("Rentang Frekuensi (Hz)")
plt.ylabel("Amplitudo Maksimum")
plt.title("Histogram Amplitudo Maksimum untuk Kejang dan No Kejang")
plt.xticks(rotation=90)
plt.legend()
plt.show()

# Visualisasi rasio amplitudo kejang terhadap no kejang
plt.figure(figsize=(12, 6))
plt.bar(merged_df["Rentang Frekuensi"], merged_df["Rasio Amplitudo"], color='purple', alpha=0.7)
plt.xlabel("Rentang Frekuensi (Hz)")
plt.ylabel("Rasio Amplitudo Kejang / No Kejang")
plt.title("Rasio Amplitudo Kejang terhadap No Kejang di Berbagai Rentang Frekuensi")
plt.xticks(rotation=90)
plt.show()

# Menampilkan statistik deskriptif
print("Statistik Deskriptif Amplitudo Kejang:")
print(kejang["Amplitudo Maksimum"].describe())
print("\nStatistik Deskriptif Amplitudo No Kejang:")
print(no_kejang["Amplitudo Maksimum"].describe())
print("\nStatistik Deskriptif Rasio Amplitudo:")
print(merged_df["Rasio Amplitudo"].describe())
```

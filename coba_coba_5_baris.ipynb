```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import Data CSV
TA29 = pd.read_csv("/content/ekstrak_chb01_05 menit terakhir.csv")
TA29
```

```python
# Mengubah semua kolom kecuali 'Time' dari format eksponensial ke desimal
columns_to_convert = TA29.columns.drop('Time')  # Pastikan kolom 'Time' tidak dikonversi

# Format angka ke desimal
pd.options.display.float_format = '{:.10f}'.format

# Konversi nilai ke desimal (tidak memengaruhi tampilan asli DataFrame di memori)
TA29[columns_to_convert] = TA29[columns_to_convert].applymap(lambda x: float(f"{x:.10f}"))

# Menyimpan kembali data ke file jika diperlukan
TA29.to_csv('data_desimal.csv', index=False)

# Tampilkan beberapa baris data untuk verifikasi
print(TA29)
```

```python
# menampilkan data yang sudah dipotong dalam domain time
# Menentukan saluran EEG yang akan ditampilkan
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3',
           'C3-P3', 'P3-O1', 'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8',
           'T8-P8-0', 'P8-O2', 'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Mengambil data waktu dari kolom 'Time'
time = TA29['Time']

# Membuat plot
plt.figure(figsize=(12, 8))

# Looping untuk menampilkan masing-masing saluran EEG yang dipilih
for channel in channels:
    if channel in TA29.columns:  # Memastikan saluran ada dalam data
        plt.plot(time, TA29[channel], label=channel)

# Menambahkan judul dan label
plt.title('EEG Data in Time Domain (During Seizure)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (µV)')
plt.legend(loc='upper right')

# Menampilkan grid dan plot
plt.grid(True)
plt.show()
```

```python
# menampilkan data yang sudah dipotong dalam domain frekuensi

# Fungsi untuk melakukan FFT dan mengembalikan frekuensi serta magnitude
def compute_fft(signal, sampling_rate):
    N = len(signal)  # Panjang sinyal
    T = 1.0 / sampling_rate  # Interval sampling
    fft_vals = np.fft.fft(signal)  # Menghitung FFT
    fft_freqs = np.fft.fftfreq(N, T)  # Frekuensi
    fft_vals = np.abs(fft_vals[:N // 2])  # Magnitude (frekuensi positif saja)
    fft_freqs = fft_freqs[:N // 2]  # Frekuensi positif saja
    return fft_freqs, fft_vals

# Sampling rate
sampling_rate = 256  # Misal 256 Hz

# Pilih beberapa atau semua saluran EEG untuk ditampilkan
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3', 'C3-P3', 'P3-O1',
            'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2',
            'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Membuat plot untuk semua saluran EEG dalam satu gambar
plt.figure(figsize=(14, 8))

for channel in channels:
    signal = TA29[channel].values  # Mengambil sinyal dari saluran EEG
    fft_freqs, fft_vals = compute_fft(signal, sampling_rate)  # Menghitung FFT
    plt.plot(fft_freqs, fft_vals, label=channel)  # Menambahkan ke plot

# Memberikan judul dan label pada plot
plt.title('FFT of All EEG Channels')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.legend(loc='upper right')  # Menampilkan legenda untuk setiap saluran
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca data (ganti dengan cara Anda membaca file jika berbentuk CSV atau DataFrame langsung)
data = pd.DataFrame({
    "FP1-F7": [5.665446e-06, 3.321123e-06, -7.619048e-06, -6.837607e-06, -4.884005e-06,
               -1.152625e-05, -9.572650e-06, 1.953602e-07, -1.191697e-05, -3.145299e-05],
    "Time": [0.964844, 0.968750, 0.972656, 0.976562, 0.980469, 0.984375, 0.988281,
             0.992188, 0.996094, 1.000000]
})

# Ekstrak waktu dan amplitudo
time = data['Time'].values
amplitude = data['FP1-F7'].values

# Hitung selisih waktu antar sampel untuk menentukan frekuensi sampling
sampling_rate = 1 / np.mean(np.diff(time))  # Frekuensi sampling (Hz)

# Terapkan FFT
fft_result = np.fft.fft(amplitude)
fft_magnitude = np.abs(fft_result)  # Magnitudo FFT
frequencies = np.fft.fftfreq(len(amplitude), d=1/sampling_rate)  # Skala frekuensi

# Ambil hanya separuh pertama (frekuensi positif)
positive_frequencies = frequencies[:len(frequencies)//2]
positive_magnitude = fft_magnitude[:len(frequencies)//2]

# Plot hasil FFT
plt.plot(positive_frequencies, positive_magnitude)
plt.title("FFT - Domain Frekuensi")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.grid()
plt.show()

# Frekuensi dominan
dominant_frequency = positive_frequencies[np.argmax(positive_magnitude)]
print(f"Frekuensi dominan: {dominant_frequency} Hz")
```

```python
import numpy as np
import pandas as pd

# Data x dari dataframe TA1 (pastikan kolomnya sesuai)
TA1 = pd.read_csv("/content/drive/MyDrive/TA/new_chb05_13.csv")  # Ganti dengan path file TA1 Anda
x = TA1['FP1-F7'].values  # Pastikan nama kolomnya adalah 'x'

# Data h dari dataframe h_result
h_result = pd.read_csv("/content/drive/MyDrive/TA/H_result.csv")  # Ganti dengan path file h_result Anda
h = h_result['H(n)'].values  # Pastikan nama kolomnya adalah 'H(n)'

# Panjang sinyal
N = len(x)
M = len(h)

# Inisialisasi hasil
Y = np.zeros(N + M - 1)

# Perhitungan Y[n] menggunakan konvolusi manual
for n in range(len(Y)):
    for k in range(n + 1):
        if k < len(x) and (n - k) < len(h):
            Y[n] += x[k] * h[n - k]

# Menampilkan seluruh hasil
for i in range(len(Y)):
    print(f"Y({i}) = {Y[i]}")
```

## 6️⃣ Grouping & Pivot

* `groupby()` → kelompokkan data lalu agregasi.
* `pivot()` dan `pivot_table()` → ubah format tabel.
* `melt()` → ubah tabel lebar ke format panjang.

````python
import pandas as pd

# DataFrame sederhana
df = pd.DataFrame({
    "Nama": ["Ari", "Ari", "Budi", "Budi", "Citra", "Citra"],
    "Bulan": ["Jan", "Feb", "Jan", "Feb", "Jan", "Feb"],
    "Penjualan": [100, 150, 200, 180, 250, 300]
})

print("=== Data Asli ===")
print(df)
````

````python
# kelompokkan data berdasarjan penjualan per nama
print(df.groupby("Nama")["Penjualan"].sum())
````

````python
# nama sebagai baris, bulan sebagai kolom penjualan
print(df.pivot(index="Nama", columns="Bulan", values="Penjualan"))
````

````python
# mirip pivot, tapi bisa agregasi
print(df.pivot_table(index="Nama", columns="Bulan", values="Penjualan", aggfunc="mean"))
````

````python
# Ubah pivot menjadi format panjang
pivot_df = df.pivot(index="Nama", columns="Bulan", values="Penjualan").reset_index()
print(pivot_df.melt(id_vars="Nama", value_name="Penjualan", var_name="Bulan"))
````

## 8️⃣ Merge & Join Data

* `merge()` → gabungkan seperti SQL join.
* `concat()` → gabungkan vertikal/horizontal.
* `join()` → gabung berdasarkan index.


````python
# dataframe 1 tidak boleh sama dengan dataframe 2
df1 = {
    "Id" : [1, 2, 3],
    "Nama" : ["Arsy", "Liza", "Dede"]
}

df1 = pd.DataFrame(df1)
df1
````

````python
# dataframe 2 tidak boleh sama dengan dataframe 1
df2= {
    "Id" : [1, 2, 4],
    "Nilai" : [80, 90, 70]
}

df2 = pd.DataFrame(df2)
df2
````

````python
# INNER JOIN → hanya ID yang sama
print(pd.merge(df1, df2, on="Id", how="inner"))
````

````python
# LEFT JOIN → semua dari df1, tambah data dari df2 jika cocok
# Pegangannya adalah tabel kiri (DataFrame pertama).
# Semua baris dari tabel kiri pasti ikut hasil.
# Data dari tabel kanan hanya ikut kalau kuncinya cocok.
# Kalau kunci nggak cocok, nilai dari tabel kanan diisi NaN.


print(pd.merge(df1, df2, on="Id", how = "left"))
````

````python
# RIGHT JOIN → semua dari df2, tambah data dari df1 jika cocok
# Pegangannya adalah tabel kanan.

print(pd.merge(df1, df2, on="Id", how="right"))
````

````python
# JOIN
# Buat DataFrame dengan index
df1 = df1.set_index("Id")
df2 = df2.set_index("Id")

print(df1.join(df2, how="inner"))  # inner join by index

````

````python
# Gabung horizontal (tambah kolom)
print(pd.concat([df1, df2], axis=1))
````

## **2️⃣ Membuat Array**

* Dari Python list:

  * `np.array([1,2,3])`
* Array berisi nol:

  * `np.zeros(shape)`
* Array berisi satu:

  * `np.ones(shape)`
* Array dengan nilai tertentu:

  * `np.full(shape, value)`
* Array dengan urutan angka:

  * `np.arange(start, stop, step)`
  * `np.linspace(start, stop, num)`
* Array acak:

  * `np.random.rand()`, `np.random.randint()`, `np.random.normal()`


| Perintah                                 | Pengertian                                                                                        | Contoh Kode                        | Contoh Bentuk Output                                |
| ---------------------------------------- | ------------------------------------------------------------------------------------------------- | ---------------------------------- | --------------------------------------------------- |
| **`np.array([1,2,3])`**                  | Membuat array dari list Python.                                                                   | `np.array([1, 2, 3])`              | `[1 2 3]`                                           |
| **`np.zeros(shape)`**                    | Membuat array berisi **0** dengan ukuran `shape`.                                                 | `np.zeros((2, 3))`                 | <pre>\[\[0. 0. 0.]\n \[0. 0. 0.]]</pre>             |
| **`np.ones(shape)`**                     | Membuat array berisi **1** dengan ukuran `shape`.                                                 | `np.ones((3, 2))`                  | <pre>\[\[1. 1.]\n \[1. 1.]\n \[1. 1.]]</pre>        |
| **`np.full(shape, value)`**              | Membuat array dengan nilai tertentu.                                                              | `np.full((2, 2), 7)`               | <pre>\[\[7 7]\n \[7 7]]</pre>                       |
| **`np.arange(start, stop, step)`**       | Membuat array dengan urutan angka dari `start` ke `stop` (tidak termasuk `stop`), langkah `step`. | `np.arange(0, 10, 2)`              | `[0 2 4 6 8]`                                       |
| **`np.linspace(start, stop, num)`**      | Membuat array dengan **jumlah nilai tertentu** antara `start` dan `stop`.                         | `np.linspace(0, 1, 5)`             | `[0.   0.25 0.5  0.75 1.]`                          |
| **`np.random.rand(shape)`**              | Membuat array acak **uniform** di \[0, 1).                                                        | `np.random.rand(2, 3)`             | <pre>\[\[0.52 0.77 0.11]\n \[0.85 0.29 0.43]]</pre> |
| **`np.random.randint(low, high, size)`** | Membuat array acak bilangan bulat dari `low` hingga `high-1`.                                     | `np.random.randint(1, 10, (3, 3))` | <pre>\[\[4 8 3]\n \[2 9 7]\n \[1 5 6]]</pre>        |
| **`np.random.normal(loc, scale, size)`** | Membuat array acak dari distribusi normal (mean=`loc`, std dev=`scale`).                          | `np.random.normal(0, 1, (2, 2))`   | <pre>\[\[ 0.12 -0.88]\n \[ 1.45  0.33]]</pre>       |

### **4️⃣ `np.polynomial` → Operasi Polinomial**

Berguna untuk membuat, menghitung, dan memanipulasi polinomial.

* **`Polynomial()`** → membuat objek polinomial
* **`.roots`** → mencari akar polinomial
* **`.deriv()`** → turunan polinomial
* **`.integ()`** → integral polinomial

````python
from numpy.polynomial import Polynomial

# Membuat polinomial p(x) = 2x² + 3x + 1
p = Polynomial([1, 3, 2])
print("p(x):", p)

# Nilai polinomial di x=2 -> semua nilai x itu 2
print("p(2) =", p(2))

# Akar polinomial -> gunain rumus kuadrat
print("Roots:", p.roots())

# Turunan
print("Derivative:", p.deriv())

# Integral
print("Integral:", p.integ())
````


**1️⃣ Membuat polinomial**

```python
p = Polynomial([1, 3, 2])
```

Artinya:

$$
p(x) = 1 + 3x + 2x^2
$$

Kenapa urutannya dari konstanta dulu? Karena `Polynomial([a0, a1, a2])` berarti:

$$
a_0 + a_1 x + a_2 x^2
$$

Jadi **bukan** urut dari pangkat tertinggi ke terendah seperti biasa di matematika.

---

**2️⃣ Nilai polinomial di $x = 2$**

$$
p(2) = 1 + 3(2) + 2(2^2)
$$

$$
p(2) = 1 + 6 + 2(4)
$$

$$
p(2) = 1 + 6 + 8 = 15
$$

---

**3️⃣ Akar polinomial**

Kita mau selesaikan:

$$
1 + 3x + 2x^2 = 0
$$

Urutkan seperti biasa:

$$
2x^2 + 3x + 1 = 0
$$

Gunakan rumus kuadrat:

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

Dengan $a = 2, b = 3, c = 1$:

$$
x = \frac{-3 \pm \sqrt{9 - 8}}{4}
$$

$$
x = \frac{-3 \pm 1}{4}
$$

Hasil:

* $x = \frac{-3 + 1}{4} = \frac{-2}{4} = -0.5$
* $x = \frac{-3 - 1}{4} = \frac{-4}{4} = -1$

Jadi:

$$
\text{Roots} = [-1, -0.5]
$$

---

**4️⃣ Turunan**

Dari:

$$
p(x) = 1 + 3x + 2x^2
$$

Turunannya:

* Turunan dari $1$ = 0
* Turunan dari $3x$ = 3
* Turunan dari $2x^2$ = $4x$

Jadi:

$$
p'(x) = 3 + 4x
$$

Dalam bentuk `Polynomial`:

$$
\text{Polynomial}([3, 4])
$$

---

**5️⃣ Integral**

Integral tak tentu:

* Integral dari $1$ = $x$
* Integral dari $3x$ = $1.5x^2$
* Integral dari $2x^2$ = $\frac{2}{3}x^3$

Jadi:

$$
\int p(x) dx = x + 1.5x^2 + \frac{2}{3}x^3 + C
$$

Jika `Polynomial` default, $C = 0$.

---

````python
import numpy as np

def fft_manual(x):
    N = len(x)

    # Basis rekursi
    if N <= 1:
        return x

    # Pecah data jadi genap & ganjil
    even = fft_manual(x[0::2])
    odd  = fft_manual(x[1::2])

    # Twiddle factor
    faktor = np.exp(-2j * np.pi * np.arange(N) / N)

    # Gabungkan hasil
    return np.concatenate([
        even + faktor[:N // 2] * odd,
        even - faktor[:N // 2] * odd
    ])

# Data contoh
data = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=complex)

# Jalankan FFT manual
hasil_manual = fft_manual(data)
print("FFT Manual:", hasil_manual)

# Bandingkan dengan FFT bawaan numpy
print("FFT NumPy :", np.fft.fft(data))
````

Kalau bagian **basis rekursi** ini:

```python
if N <= 1:
    return x
```

dihapus, maka fungsi akan **terjebak dalam pemanggilan rekursi tak berujung** dan akhirnya menyebabkan **`RecursionError`** di Python.

---

**Kenapa?**

FFT rekursif itu bekerja dengan cara **membagi masalah besar menjadi masalah kecil** sampai masalahnya **tidak bisa dibagi lagi**.

* Misalnya data `x` panjangnya **8**, fungsi akan memanggil dirinya sendiri dengan panjang **4**, lalu **2**, lalu **1**.
* Kalau **N = 1**, sebenarnya sudah tidak ada yang dihitung (FFT dari 1 titik adalah dirinya sendiri).
* **Tanpa `if N <= 1: return x`**, Python akan terus mencoba memanggil `fft_manual` dengan data yang bahkan bisa kosong (`[]`) dan **tidak pernah berhenti**.

---

📌 **Contoh analogi sederhana**:
Bayangin kamu punya tugas: "Bagi kue jadi dua sampai ukurannya kecil banget".
Kalau tidak ada aturan **"berhenti kalau ukurannya 1"**, kamu akan **terus membelah** kue sampai *tak terhingga*, padahal di dunia nyata kuenya sudah habis.

---

💡 **Kesimpulan**:
Baris `if N <= 1: return x` adalah **syarat berhenti** dalam algoritma rekursif FFT.
Kalau dihapus, algoritma tidak punya titik akhir, sehingga program akan **error karena rekursi terlalu dalam**.

---

Kalau mau, aku bisa bikin **simulasi tabel langkah rekursinya** supaya kamu bisa jelasin ini dengan gampang pas sidang.
Mau aku bikinin?


Oke, kita bongkar baris ini:

```python
faktor = np.exp(-2j * np.pi * np.arange(N) / N)
```

---

 1️⃣ **Makna dalam teori FFT**

Baris itu menghitung **Twiddle Factor**

$$
W_N^k = e^{-\frac{2\pi i k}{N}}
$$

* **$N$** = panjang data
* **$k$** = indeks frekuensi (0, 1, 2, …, N-1)
* **$i$** atau **$j$** = unit imajiner ($i^2 = -1$)
* **$e^{i\theta}$** = bentuk Euler: $e^{i\theta} = \cos\theta + i\sin\theta$

---

 2️⃣ **Hubungan kode dengan rumus**

`np.arange(N)` → menghasilkan $k = 0, 1, 2, …, N-1$
`-2j * np.pi * np.arange(N) / N` → menghasilkan $-\frac{2\pi i k}{N}$
`np.exp(...)` → menghitung $e^{(\text{...})}$

Jadi kodenya persis seperti rumus:

$$
faktor[k] = e^{-\frac{2\pi i k}{N}}, \quad k = 0,1,\dots,N-1
$$

---

 3️⃣ **Bentuk trigonometri (dari rumus Euler)**

Dengan **Rumus Euler**:

$$
W_N^k = \cos\left(\frac{2\pi k}{N}\right) - i \sin\left(\frac{2\pi k}{N}\right)
$$

---

✅ **Kesimpulan**:
Kode tersebut membuat **array** Twiddle Factor:

$$
\big[ e^{-\frac{2\pi i \cdot 0}{N}},\ e^{-\frac{2\pi i \cdot 1}{N}},\ e^{-\frac{2\pi i \cdot 2}{N}},\dots,\ e^{-\frac{2\pi i (N-1)}{N}} \big]
$$

````python
import numpy as np
from scipy.signal import firwin

# BAND PASS FILTER

# Parameter
fs = 256       # Sampling rate (Hz)
f_low = 1      # Frekuensi batas bawah (Hz)
f_high = 30    # Frekuensi batas atas (Hz)
numtaps = 10   # Jumlah koefisien filter (tap) — diperbesar biar responnya lebih tajam
nyquist = fs/2

# Normalisasi cutoff ke satuan Nyquist (0 - 1)
cutoff_norm = [f_low / (nyquist), f_high / (nyquist)]

# Hitung koefisien filter Band Pass
taps = firwin(numtaps, cutoff_norm, window='hamming', pass_zero='bandpass')

# Tampilkan hasil koefisien
print("Koefisien filter FIR Band Pass:")
print(taps)
````

````python
def gini_index(counts):

    total = sum(counts)
    if total == 0:
        return 0

    gini = 1.0
    for c in counts:
        p = c / total   # proporsi kelas
        gini -= p ** 2  # kurangi dengan p^2

    return gini

print(gini_index([5, 3, 2]))  # output: 0.66
print(gini_index([10, 0]))    # output: 0.0 (pure, hanya 1 kelas)
print(gini_index([5, 5]))     # output: 0.5 (dua kelas seimbang)
````

````python
def gini_split(groups):

    # total semua sampel di semua grup
    total = sum(sum(g) for g in groups)
    if total == 0:
        return 0

    gini_split_value = 0
    for g in groups:
        size = sum(g)  # jumlah sampel di grup ini
        if size == 0:
            continue
        # hitung gini index grup ini
        gini_g = gini_index(g)
        # tambahkan ke total dengan bobot proporsional
        gini_split_value += (size / total) * gini_g

    return gini_split_value

left = [5, 3]      # grup kiri (kelas A=5, B=3)
right = [4, 2, 1]  # grup kanan (kelas A=4, B=2, C=1)

print("Gini Split:", gini_split([left, right]))
````

````python
import numpy as np

def manual_accuracy(cm):
    # Ambil nilai TN, FP, FN, TP dari confusion matrix
    tn = cm[0][0]   # baris 0, kolom 0
    fp = cm[0][1]   # baris 0, kolom 1
    fn = cm[1][0]   # baris 1, kolom 0
    tp = cm[1][1]   # baris 1, kolom 1

    total = tn + fp + fn + tp
    accuracy = (tp + tn) / total
    return accuracy

# Contoh pakai
cm = np.array([[50, 10],
               [5, 35]])

print("Accuracy:", manual_accuracy(cm))
````

````python
import numpy as np

def recall_positif(cm):
    # Ambil nilai TN, FP, FN, TP dari confusion matrix
    tn = cm[0][0]   # baris 0, kolom 0
    fp = cm[0][1]   # baris 0, kolom 1
    fn = cm[1][0]   # baris 1, kolom 0
    tp = cm[1][1]   # baris 1, kolom 1

    positif = tp + fn
    recall = tp / positif
    return recall

# Contoh pakai
cm = np.array([[50, 10],
               [5, 35]])

print("Recall:", recall_positif(cm))
````

````python
import numpy as np

def recall_negatif(cm):
    # Ambil nilai TN, FP, FN, TP dari confusion matrix
    tn = cm[0][0]   # baris 0, kolom 0
    fp = cm[0][1]   # baris 0, kolom 1
    fn = cm[1][0]   # baris 1, kolom 0
    tp = cm[1][1]   # baris 1, kolom 1

    negatif = tn + fp
    recall = tn / negatif
    return recall

# Contoh pakai
cm = np.array([[50, 10],
               [5, 35]])

print("Recall:", recall_negatif(cm))
````

````python
import numpy as np

def presisi_positif(cm):
    # Ambil nilai TN, FP, FN, TP dari confusion matrix
    tn = cm[0][0]   # baris 0, kolom 0
    fp = cm[0][1]   # baris 0, kolom 1
    fn = cm[1][0]   # baris 1, kolom 0
    tp = cm[1][1]   # baris 1, kolom 1

    positif = tp + fp
    recall = tp / positif
    return recall

# Contoh pakai
cm = np.array([[50, 10],
               [5, 35]])

print("Recall:", presisi_positif(cm))
````

````python
import numpy as np

def presisi_negatif(cm):
    # Ambil nilai TN, FP, FN, TP dari confusion matrix
    tn = cm[0][0]   # baris 0, kolom 0
    fp = cm[0][1]   # baris 0, kolom 1
    fn = cm[1][0]   # baris 1, kolom 0
    tp = cm[1][1]   # baris 1, kolom 1

    negatif = tn + fn
    recall = tn / negatif
    return recall

# Contoh pakai
cm = np.array([[50, 10],
               [5, 35]])

print("Recall:", presisi_negatif(cm))
````

````python
def f1_score(cm):
    p = presisi_positif(cm)
    r = recall_positif(cm)
    return (2 * p * r) / (p + r)

print("F1 Score :", f1_score(cm))
````

````python
from scipy.signal import firwin

fs = 256
low = 1
high = 30
numtaps = 10
nyquist = fs/2

cutoff_norm = [low/nyquist, high/nyquist]

taps = firwin(numtaps, cutoff_norm, window='hamming', pass_zero='bandpass')
taps
````

````python
from scipy.signal import lfilter

filtered = lfilter(taps, 1, signal)
filtered
````

````python
import numpy as np

def fft_manual(x):
    N = len(x)

    # Basis rekursi
    if N <= 1:
        return x

    # Pecah data jadi genap & ganjil
    even = fft_manual(x[0::2])
    odd  = fft_manual(x[1::2])

    # Twiddle factor
    faktor = np.exp(-2j * np.pi * np.arange(N) / N)

    # Gabungkan hasil
    return np.concatenate([
        even + faktor[:N // 2] * odd,
        even - faktor[:N // 2] * odd
    ])


# Jalankan FFT manual
hasil_manual = fft_manual(filtered)
print("FFT Manual:", hasil_manual)

# Bandingkan dengan FFT bawaan numpy
hasil_numpy = np.fft.fft(filtered)
print("FFT NumPy :", hasil_numpy)
````

````python
import numpy as np

# HAMMING WINDOW
def h(n, f1, f2, N):
    term1 = 2 * f1 * (np.sin(n * 2 * np.pi * f1) / (n * 2 * np.pi * f1))
    term2 = 2 * f2 * (np.sin(n * 2 * np.pi * f2) / (n * 2 * np.pi * f2))

    window = 0.54 + 0.46 * np.cos((2 * np.pi * n) / (N + 1))

    return (term2 - term1) * window

# cara penggunaan
f1 = 0.0078125
f2 = 0.234375
N = 10

print(h(1, f1, f2, N))
print(h(2, f1, f2, N))
````

````python
import numpy as np

# Fungsi konvolusi manual
def konvolusi(x, h):
    """
    x : sinyal input (list atau numpy array)
    h : respon impuls filter (list atau numpy array)
    return : hasil konvolusi (numpy array)
    """
    N = len(x)
    M = len(h)
    y = np.zeros(N + M - 1)  # panjang hasil konvolusi

    # Proses konvolusi manual
    for n in range(N + M - 1):
        total = 0
        for k in range(N):
            if 0 <= n-k < M:
                total += x[k] * h[n-k]
        y[n] = total

    return y

# ==== Contoh penggunaan ====
# Sinyal input
x = [1, 2, 3, 4]
# Respon impuls filter sederhana (misal low-pass FIR kecil)
h = [0.25, 0.5, 0.25]

y = konvolusi(x, h)
print("Hasil konvolusi:", y)
````

````python
import numpy as np

# Data input (N=4)
data = np.array([1, 2, 3, 4])

# Pisahkan even dan odd
even = data[[0, 2]]  # index genap
odd  = data[[1, 3]]  # index ganjil

# ------------------------
# FFT untuk even (N=2)
# ------------------------
a0 = even[0] + even[1]
a1 = even[0] - even[1]

# ------------------------
# FFT untuk odd (N=2)
# ------------------------
b0 = odd[0] + odd[1]
b1 = odd[0] - odd[1]

# ------------------------
# Combine hasil (N=4)
# ------------------------
W0 = np.exp(-2j * np.pi * 0 / 4)
W1 = np.exp(-2j * np.pi * 1 / 4)

K0 = a0 + W0 * b0
K1 = a1 + W1 * b1
K2 = a0 - W0 * b0
K3 = a1 - W1 * b1

# Hasil akhir
hasil_fft = [K0, K1, K2, K3]
print("Hasil FFT manual (N=4):", hasil_fft)

# Bandingkan dengan numpy.fft
print("FFT numpy:", np.fft.fft(data))
````

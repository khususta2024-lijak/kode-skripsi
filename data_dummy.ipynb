```python
import numpy as np
import pandas as pd

# Konstanta yang digunakan
pi = np.pi

# Inisialisasi list untuk menyimpan hasil
results = []

# Loop untuk n dari 1 hingga 868
for n in range(1, 869):
    # Bagian pertama dari persamaan
    factor1 = 2 * 0.1191
    sin1 = np.sin((n * 2 * pi * 0.1191))
    divisor1 = (n * 2 * pi * 0.1191)
    term1 = sin1 / divisor1

    # Bagian kedua dari persamaan
    factor2 = 2 * 0.0029
    sin2 = np.sin((n * 2 * pi * 0.0029))
    divisor2 = (n * 2 * pi * 0.0029)
    term2 = sin2 / divisor2

    # Bagian ketiga (dalam kurung besar) dari persamaan
    cos_term = np.cos((2 * pi * n) / 1735)
    inner_bracket = 0.54 + 0.46 * cos_term

    # Menghitung nilai akhir H(n)
    H_n = factor1 * term1 - factor2 * term2 * inner_bracket

    # Menyimpan hasil ke dalam list
    results.append({"n": n, "H(n)": H_n})

# Membuat dataframe dari hasil
df = pd.DataFrame(results)

# Menampilkan dataframe
print(df)

# (Opsional) Menyimpan dataframe ke file CSV
df.to_csv("/content/drive/MyDrive/TA/H_result.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# Membaca data dari file
TA1_sesudah = pd.read_csv("/content/drive/MyDrive/TA/Y_result_manual_30,25.csv")

# Mengambil data x(n)
x_n = TA1_sesudah['Y(n)'].values
N = len(x_n)  # Panjang data

# Fungsi untuk menghitung FFT manual
def fft_manual(x_n):
    N = len(x_n)
    fft_result = []  # Menyimpan hasil FFT untuk setiap k

    for k in range(N):
        X_k = 0  # Inisialisasi X(k)
        for n in range(N):
            # Menghitung X(k) sesuai rumus
            angle = -2j * np.pi * k * n / N
            X_k += x_n[n] * np.exp(angle)
        fft_result.append(X_k)

    return np.array(fft_result)

# Menghitung FFT manual
fft_result = fft_manual(x_n)

# Membuat DataFrame hasil
fft_df = pd.DataFrame({
    'Frequency': np.fft.fftfreq(N, d=1),  # Frekuensi
    'Real': fft_result.real,  # Bagian real
    'Imaginary': fft_result.imag,  # Bagian imajiner
    'Amplitude': np.abs(fft_result),  # Amplitudo
    'Phase': np.angle(fft_result)  # Fase
})

# Menampilkan beberapa hasil
print(fft_df.head())

# Menyimpan hasil ke file (opsional)
fft_df.to_csv("/content/drive/MyDrive/TA/fft_manual_result.csv", index=False)
```

```python
import numpy as np

# Data x(n)
x_n = [9.691e-6, 1.700e-5, 2.007, 2.008]
N = 100.097  # Panjang data yang digunakan dalam perhitungan
k_values = [0, 1, 2, 3]  # Nilai k yang ingin dihitung

# Fungsi untuk menghitung FFT manual
def fft_manual(x_n, N, k_values):
    results = {}

    for k in k_values:
        X_k = 0  # Inisialisasi X(k)
        details = []  # Untuk menyimpan detail proses perhitungan

        for n, x in enumerate(x_n):
            angle = -2j * np.pi * k * n / N  # Menghitung sudut
            exponential = np.exp(angle)  # Menghitung nilai eksponensial
            term = x * exponential  # Kontribusi term dari x(n) * e^(...)
            X_k += term

            # Menyimpan detail
            details.append({
                "n": n,
                "x_n": x,
                "angle": angle,
                "e^(...j)": exponential,
                "term": term
            })

        results[k] = {
            "X_k": X_k,
            "details": details
        }

    return results

# Menghitung FFT manual
fft_results = fft_manual(x_n, N, k_values)

# Menampilkan hasil perhitungan
for k, result in fft_results.items():
    print(f"\nHasil untuk X({k}):")
    print(f"X({k}) = {result['X_k']}")
    print("Detail perhitungan:")
    for detail in result['details']:
        print(
            f"  n={detail['n']}, x(n)={detail['x_n']}, angle={detail['angle']}, "
            f"e^(...j)={detail['e^(...j)']}, term={detail['term']}"
        )
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan FFT dan mengembalikan frekuensi serta magnitudo
def compute_fft(signal, sampling_rate):
    N = len(signal)  # Panjang sinyal
    T = 1.0 / sampling_rate  # Interval sampling
    fft_vals = np.fft.fft(signal)  # Menghitung FFT
    fft_freqs = np.fft.fftfreq(N, T)  # Frekuensi
    fft_vals = np.abs(fft_vals)  # Magnitudo (tidak hanya frekuensi positif)
    return fft_freqs, fft_vals

# Sampling rate
sampling_rate = 256  # Misal 256 Hz

# Pilih beberapa atau semua saluran EEG untuk ditampilkan
channels = ['FP1-F7', 'F7-T7', 'T7-P7', 'P7-O1', 'FP1-F3', 'F3-C3', 'C3-P3', 'P3-O1',
            'FP2-F4', 'F4-C4', 'C4-P4', 'P4-O2', 'FP2-F8', 'F8-T8', 'T8-P8-0', 'P8-O2',
            'FZ-CZ', 'CZ-PZ', 'P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1']

# Membuat plot untuk semua saluran EEG dalam satu gambar
plt.figure(figsize=(14, 8))

for channel in channels:
    signal = data1[channel].values  # Mengambil sinyal dari saluran EEG
    fft_freqs, fft_vals = compute_fft(signal, sampling_rate)  # Menghitung FFT
    plt.plot(fft_freqs, fft_vals, label=channel)  # Menambahkan ke plot

# Memberikan judul dan label pada plot
plt.title('FFT of All EEG Channels')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.legend(loc='upper right')  # Menampilkan legenda untuk setiap saluran
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Fungsi untuk melakukan FFT dan mengembalikan frekuensi serta magnitude
def compute_fft(signal, sampling_rate):
    N = len(signal)  # Panjang sinyal
    T = 1.0 / sampling_rate  # Interval sampling
    fft_vals = np.fft.fft(signal)  # Menghitung FFT
    fft_freqs = np.fft.fftfreq(N, T)  # Frekuensi
    fft_vals = np.abs(fft_vals[:N // 2])  # Magnitude (frekuensi positif saja)
    fft_freqs = fft_freqs[:N // 2]  # Frekuensi positif saja
    return fft_freqs, fft_vals

# Sampling rate
sampling_rate = 256  # Misal 256 Hz

# Pilih beberapa atau semua saluran EEG untuk ditampilkan
channels = ['FP1-F7']

# DataFrame untuk menyimpan hasil FFT
fft_results_df = pd.DataFrame()

for channel in channels:
    signal = CHANNEL_2AJA[channel].values  # Mengambil sinyal dari saluran EEG
    fft_freqs, fft_vals = compute_fft(signal, sampling_rate)  # Menghitung FFT

    # Menambahkan hasil FFT ke DataFrame
    fft_results_df[channel + ' Freq'] = fft_freqs
    fft_results_df[channel + ' Magnitude'] = fft_vals

# Menampilkan DataFrame hasil FFT
print("FFT Results DataFrame:")
print(fft_results_df)

# Membuat plot interaktif menggunakan plotly
fig = go.Figure()

for channel in channels:
    fig.add_trace(go.Scatter(x=fft_results_df[channel + ' Freq'],
                             y=fft_results_df[channel + ' Magnitude'],
                             mode='lines', name=channel))

# Memberikan judul dan label pada plot
fig.update_layout(
    title='FFT of All EEG Channels',
    xaxis_title='Frequency (Hz)',
    yaxis_title='Magnitude',
    showlegend=True,
    template='plotly_dark'
)

# Menambahkan fitur interaktif yang menampilkan nilai saat kursor berada di atas garis
fig.update_traces(mode='lines+markers', hoverinfo='x+y')

# Menampilkan grafik interaktif
fig.show()
```

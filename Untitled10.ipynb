```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/TA/HANNING BALANCE FFT YANG DIPISAH KEJANG DAN TIDAK KEJANG.csv")
df
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt


# =====================
# 2. Pisahkan fitur dan target
# =====================
X = df.drop(columns=["seizure", "frekuensi"])
y = df["seizure"]

# Encode target (kejang/tidak kejang -> 1/0)
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# =====================
# 3. Train Random Forest
# =====================
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
rf.fit(X, y_encoded)

# =====================
# 4. Ambil feature importance
# =====================
importances = rf.feature_importances_
feature_names = X.columns

# Buat DataFrame ranking
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print(importance_df)

# =====================
# 5. Visualisasi
# =====================
plt.figure(figsize=(10,6))
importance_df.head(20).plot(kind="barh", x="Feature", y="Importance", legend=False)
plt.title("Top 20 Feature Importances - Random Forest")
plt.gca().invert_yaxis()
plt.show()
```

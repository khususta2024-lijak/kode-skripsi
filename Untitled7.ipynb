```python
def h(f1, f2, N):
  n = np.arange(1, N+1)

  term1 = (2 * f1) * (np.sin(2 * n * np.pi * f1) / (2 * n * np.pi * f1))
  term2 = (2 * f2) * (np.sin(2 * n * np.pi * f2) / (2 * n * np.pi * f2))

  window = 0.42 + 0.5 * (np.cos(2 * np.pi * n)/(N)) + 0.08 * (np.cos(4 * np.pi * n)/(N))

  return (term2-term1) * window

f1 = 1 / 128
f2 = 50 / 128
N = 20

tap = h(f1, f2, N)
print(tap)
```

```python
def konvolusi(x, h):
  N = len(x)
  M = len(h)

  panjang = N + M - 1
  y = [0] * panjang

  for n in range(panjang):
    for k in range(N):
      if 0 <= n-k < M:
        y[n] += x[k] * h[n-k]
  return np.array(x[:N])

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])

filter = konvolusi(x, tap)
filter
```

```python
# konvolusi manual
def konvolusi(x, h):
  N = len(x)
  M = len(h)

  panjang = N + M - 1
  y = [0] * panjang

  for n in range(panjang):
    for k in range(N):
      if 0 <= n-k < M:
        y[n] += x[k] * h[n-k]
  return np.array(x[:N])

konvo = konvolusi(arr, impuls_respon)
konvo
```

```python
# fft manual pake data sendiri
def fft_manual(x):
  N = len(x)
  if N <= 1:
    return x

  even = fft_manual(x[0::2])
  odd = fft_manual(x[1::2])

  faktor = np.exp (-2j * np.pi * np.arange(N) / (N))

  return np.concatenate([
      even + faktor[:N // 2] * odd,
      even - faktor[:N // 2] * odd
  ])

arr1 = np.array([1, 2, 3, 4])
fft1 = fft_manual(arr1)
print('fft1 = ', fft1)

arr2 = np.array([1, 2, 3, 4, 5, 6, 7, 8])
fft2 = fft_manual(arr2)
print('fft2 = ', fft2)
```

```python
# standar deviasi
def standar_deviasi(data):
  N = len(data)
  mean = sum(data) / N
  jumlah_kuadrat = 0

  for x in data:
    selisih = x - mean
    kuadrat = selisih ** 2
    jumlah_kuadrat += selisih ** 2

  rata_kuadrat = jumlah_kuadrat / N
  std = np.sqrt(rata_kuadrat)
  return std


data = [70, 75, 80, 85, 90]
hasil = standar_deviasi(data)
print(hasil)
```

```python
# fungsi filtering
import numpy as np
from scipy.signal import firwin

# BAND PASS FILTER

# Parameter
fs = 256       # Sampling rate (Hz)
f_low = 1      # Frekuensi batas bawah (Hz)
f_high = 10    # Frekuensi batas atas (Hz)
numtaps = 10   # Jumlah koefisien filter (tap) â€” diperbesar biar responnya lebih tajam
nyquist = fs/2

# Normalisasi cutoff ke satuan Nyquist (0 - 1)
cutoff_norm = [f_low / (nyquist), f_high / (nyquist)]

# Hitung koefisien filter Band Pass
taps = firwin(numtaps, cutoff_norm, window='hamming', pass_zero=False)

# Tampilkan hasil koefisien
print("Koefisien filter FIR Band Pass:")
print(taps)
```

```python

# terapin filtering
# Terapkan filter FIR ke sinyal array
from scipy.signal import lfilter

y = lfilter(taps, 1.0, TA) #1.0 artinya tidak ada komponen feedback(murni FIR, bukan IIR)

print("Sinyal asli:")
print(TA)

print("\nSinyal setelah filtering:")
print(y)
```

## latihan max min mean stdv energi

```python
x = np.array([1, 2, 2, 5, 5, 9, 7])

def cari_mean(x):
    total = 0
    for nilai in x:
        total += nilai
    return total / len(x)

print(cari_mean(x))
```

```python
x = np.array([1, 2, 2, 5, 5, 9, 7])

def cari_max(x):
    maksimum = x[0]
    for nilai in x:
        if nilai > maksimum:
            maksimum = nilai
    return maksimum

print(cari_max(x))
```

```python
x = np.array([1, 2, 2, 5, 5, 9, 7])

def cari_min(x):
    minimum = x[0]
    for nilai in x:
        if nilai < minimum:
            minimum = nilai
    return minimum

print(cari_min(x))
```

```python
x = np.array([1, 2, 2, 5, 5, 9, 7])

def cari_stdv(x):
    mean = cari_mean(x)
    total = 0
    for nilai in x:
        total += (nilai - mean) ** 2
    return (total / len(x)) ** 0.5

print(cari_stdv(x))
```

```python
x = np.array([1, 2, 2, 5, 5, 9, 7])

def cari_energi(x):
    total = 0
    for nilai in x:
        total += abs(nilai) ** 2
    return total

print(cari_energi(x))
```

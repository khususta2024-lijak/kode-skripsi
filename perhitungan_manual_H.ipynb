```python
# cari impulse respon dulu
import numpy as np
import plotly.graph_objects as go

# Konstanta yang digunakan
pi = np.pi

# Rentang nilai n
n_values = np.arange(1, 847)

# Menghitung H[n] untuk setiap n
H_values = []
for n in n_values:
    # Bagian pertama dari persamaan
    factor1 = 2 * 0.0078
    sin1 = np.sin((n * 2 * pi * 0.0078))
    divisor1 = (n * 2 * pi * 0.0078)
    term1 = sin1 / divisor1

    # Bagian kedua dari persamaan
    factor2 = 2 * 0.2343
    sin2 = np.sin((n * 2 * pi * 0.2343))
    divisor2 = (n * 2 * pi * 0.2343)
    term2 = sin2 / divisor2

    # Bagian ketiga (dalam kurung besar) dari persamaan
    cos_term = np.cos((2 * pi * n) / 847)
    inner_bracket = 0.54 + 0.46 * cos_term

    # Menghitung nilai H[n]
    H_n = factor1 * term1 - factor2 * term2 * inner_bracket
    H_values.append(H_n)

# Menampilkan hasil H[n]
for n, H_n in zip(n_values, H_values):
    print(f"H({n}) = {H_n:.4f}")

# Membuat plot menggunakan plotly.graph_objects
fig = go.Figure()

# Menambahkan garis sinyal
fig.add_trace(go.Scatter(
    y=H_values,
    mode='lines',
    name='Sinyal',
    hovertemplate='Index: %{x}<br>Value: %{y:.8f}<extra></extra>'
))

# Mengatur layout
fig.update_layout(
    title='Visualisasi Sinyal dengan Interaktivitas',
    xaxis_title='Index',
    yaxis_title='Amplitude',
    hovermode='x',
    template='plotly'
)

# Menampilkan plot
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Ambil data dari kolom 'FP1-F7' di df_filtered
x = df_filtered['FP1-F7'].values  # Menambahkan .values untuk mengubah ke array NumPy

h = H_values  # Sesuaikan dengan data h Anda

# Panjang sinyal
N = len(x)
M = len(h)

# Inisialisasi hasil
Y = np.zeros(N + M - 1)

# Perhitungan Y[n]
for n in range(len(Y)):
    for k in range(n + 1):
        if k < len(x) and (n - k) < len(h):
            Y[n] += x[k] * h[n - k]

# Menampilkan 5 baris pertama hasil Y[n]
for i in range(min(5, len(Y))):  # Menampilkan hanya 5 baris pertama
    print(f"Y({i}) = {Y[i]}")

# Membuat plot menggunakan plotly.graph_objects
fig = go.Figure()

# Menambahkan garis sinyal
fig.add_trace(go.Scatter(
    y=Y,
    mode='lines',
    name='Sinyal',
    hovertemplate='Index: %{x}<br>Value: %{y:.8f}<extra></extra>'
))

# Mengatur layout
fig.update_layout(
    title='Visualisasi Sinyal dengan Interaktivitas',
    xaxis_title='Index',
    yaxis_title='Amplitude',
    hovermode='x',
    template='plotly'
)

# Menampilkan plot
fig.show()
```

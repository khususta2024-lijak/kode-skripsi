```python
import numpy as np
import scipy.signal as signal
import pandas as pd

# Load data EEG
TA_1 = pd.read_csv("/content/drive/MyDrive/TA/FILTERING/chb05_06.csv")

# Sampling rate EEG
sampling_rate = 256

# Fungsi untuk menerapkan FIR Band-Pass Filter dengan Hamming window
def apply_fir_bandpass(signal_data, lowcut, highcut, sampling_rate, num_taps=101):
    nyquist = 0.5 * sampling_rate  # Nyquist frequency
    low = lowcut / nyquist
    high = highcut / nyquist

    # Buat koefisien filter dengan Hamming window
    fir_coeff = signal.firwin(num_taps, [low, high], pass_zero=False, window="hamming", fs=sampling_rate)

    # Terapkan filter ke sinyal EEG
    filtered_signal = signal.lfilter(fir_coeff, 1.0, signal_data)

    return filtered_signal

# Pilih channel EEG untuk diproses (tanpa kolom Time & seizure)
channels = [col for col in TA_1.columns if col not in ['Time', 'seizure']]

# Buat DataFrame untuk hasil filtering
filtered_data_kejang = pd.DataFrame()
filtered_data_tidak_kejang = pd.DataFrame()

for channel in channels:
    # Filtering 1-25 Hz untuk data kejang
    filtered_data_kejang[channel] = apply_fir_bandpass(TA_1[channel].values, 1, 25, sampling_rate)

    # Filtering 30-50 Hz untuk data tidak kejang
    filtered_data_tidak_kejang[channel] = apply_fir_bandpass(TA_1[channel].values, 30, 50, sampling_rate)

# Tambahkan label
filtered_data_kejang['seizure'] = 1
filtered_data_tidak_kejang['seizure'] = 0

# Gabungkan data hasil filtering
filtered_data = pd.concat([filtered_data_kejang, filtered_data_tidak_kejang])

# Cek distribusi label
print(filtered_data['seizure'].value_counts())

# Simpan hasil filtering jika diperlukan
filtered_data.to_csv("/content/drive/MyDrive/TA/FILTERING/filtered_eeg_1-25_30-50.csv", index=False)
```
